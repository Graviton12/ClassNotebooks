### Write a Python program to sort a dictionary by key.
nums = {6: 10, 5: 20, 4: 30, 3: 40, 2: 50, 1: 60}

```python
nums = {6: 10, 5: 20, 4: 30, 3: 40, 2: 50, 1: 60}
sorted_nums=dict()
for k,v in sorted(nums.items()):
    sorted_nums[k]=v
print(sorted_nums)
```

```python
sorted(nums.items()) # Creating a List
```
